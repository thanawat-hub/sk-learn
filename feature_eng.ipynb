```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, SelectPercentile, GenericUnivariateSelect, \
                                      chi2, f_classif, mutual_info_classif
```

```python
X = load_iris(as_frame=True)
X.data
```

```python
X.target
```

```python
###
# ฟังก์ชันเหล่านี้ช่วยให้เลือกฟีเจอร์ที่มีประโยชน์ที่สุด
# สำหรับการสร้างโมเดลจำแนกประเภทที่มีประสิทธิภาพ
# ---
score_func = chi2 # วัดความสัมพันธ์ระหว่างฟีเจอร์แบบกลุ่ม (categorical) กับตัวแปรเป้าหมาย (target)
# โดยใช้สถิติไคสแควร์ (chi-square) ฟีเจอร์ที่มีค่า chi2 สูงบ่งบอกถึงความสัมพันธ์ที่
# แข็งแกร่งกับตัวแปรเป้าหมาย


# score_func = f_classif # วัดความสัมพันธ์ระหว่างฟีเจอร์แบบต่อเนื่อง (numerical) กับตัวแปรเป้าหมาย
# โดยใช้ ANOVA (Analysis of Variance) ฟีเจอร์ที่มีค่า f_classif สูงบ่งบอกถึงความสัมพันธ์ที่
# แข็งแกร่งกับตัวแปรเป้าหมาย
# score_func = mutual_info_classif # วัดการพึ่งพาอาศัยกันระหว่างฟีเจอร์แบบใดๆ กับตัวแปรเป้าหมาย
# โดยใช้วัดข้อมูลร่วม (mutual information) ฟีเจอร์ที่มีค่า mutual_info_classif สูงบ่งบอกถึง
# ความสัมพันธ์ที่แข็งแกร่งกับตัวแปรเป้าหมาย
# ---
###
```

```python
# selector = SelectKBest(score_func=score_func, k=2) ## เทียบเท่าตัวล่าง 
selector = GenericUnivariateSelect(score_func=score_func, mode='k_best', param=2)
selector.fit(X.data, X.target)
print(selector.scores_)
print(selector.get_feature_names_out())
```

```python
# selector = SelectPercentile(score_func=score_func, percentile=30)
selector = GenericUnivariateSelect(score_func=score_func, mode='percentile', param=30)
selector.fit(X.data, X.target)
print(selector.scores_)
```

```python
selector = GenericUnivariateSelect(score_func=score_func, mode='fwe', param=0.05)
selector.fit(X.data, X.target)
print(selector.scores_)
print(selector.pvalues_)
print(selector.get_feature_names_out())
```

## Exhaustive (All combinations)

```python
from itertools import combinations, chain # ต่อกันคือใช้ chain
# ต่อจากอะไร เช่น iris มี 4 อย่าง (feature) เราก็ต้องเรียก 1 เสร็จ ก็เลือก2 , 3, 4 เท่าด้านล่างครั้ง แล้วเอาแต่ละอันมาบวกกัน
```

```python
for i in range(1, 5):
    for x in combinations([0, 1, 2, 3], i): # เริ่มที่ 0
        print(x)
```

```python
n_features = X.data.shape[1]
chain.from_iterable(combinations(range(n_features), i) for i in range(1, 5)) 
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def EFS(estimator, X, Y, cv=5, verbose=False): #Exhaustive feature selection
    n_features = X.shape[1]
    subsets = chain.from_iterable(combinations(range(n_features), i) for i in range(1, 5))
    best_score = -np.inf
    best_subset = None#คงไม่น้อยกว่า อนัน อยู่แล้ว
    for i, subset in enumerate(subsets):
        subset = list(subset)
        score = cross_val_score(estimator, X.iloc[:, subset], Y, cv=cv).mean() # ถ้า score ใหม่ดีกว่า 
        # เก็บเป็นค่าใหม่ที่ดีสุด 
        if score > best_score:
            best_score = score
            best_subset = subset
        if verbose:
            print(i, score, subset)
    return X.columns[best_subset]
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)
selected = EFS(clf, X.data, X.target, cv=StratifiedKFold(5), verbose=True)
print(selected)
```
